```python
import os
import csv
import random

def leer_archivos_txt(ruta):
    datos = []
    for archivo in os.listdir(ruta):
        if archivo.endswith(".txt"):
            with open(os.path.join(ruta, archivo), 'r') as file:
                lineas = file.readlines()
                for linea in lineas:
                    partes = linea.strip().split(":")
                    if len(partes) == 2:
                        cp, localidad = partes
                        datos.append((cp, localidad))
                    else:
                        print(f"Advertencia: Se omitió una línea incorrecta en {archivo}: {linea.strip()}")
    return datos

def generar_cobertura(datos):
    datos_con_cobertura = []
    for cp, localidad in datos:
        if cp.startswith("28"):
            cobertura = "Sí"
        else:
            cobertura = "Sí" if random.random() < 0.8 else "No"  # Probabilidad de 80% de tener cobertura
        datos_con_cobertura.append((cp, localidad, cobertura))
    return datos_con_cobertura

def guardar_csv(datos, filename):
    mode = 'w' if not os.path.exists(filename) else 'a'  # Modo 'a' si el archivo ya existe
    with open(filename, mode=mode, newline='') as file:
        writer = csv.writer(file)
        if mode == 'w':
            writer.writerow(["CodigoPostal", "Localidad", "Cobertura"])
        writer.writerows(datos)

# Directorio que contiene los archivos de texto
directorio_txt = os.path.join(os.path.expanduser("~"), "Desktop", "chatbot", "codigos1220n")
# Nombre del archivo CSV
archivo_csv = os.path.join(os.path.expanduser("~"), "Desktop", "chatbot", "cobertura.csv")

# Leer archivos de texto y extraer los códigos postales y localidades
datos = leer_archivos_txt(directorio_txt)

# Generar cobertura aleatoria para cada entrada
datos_con_cobertura = generar_cobertura(datos)

# Guardar los datos en un archivo CSV
guardar_csv(datos_con_cobertura, archivo_csv)

print("CSV generado/actualizado exitosamente.")

```
